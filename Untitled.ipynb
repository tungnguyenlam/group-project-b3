Viết lại hết: từ SegDataset -> test với yolov8s -> tìm cách cho hết về cùng dạng -> viết EvalSeg -> test tiếp

```python
series_list= ["Arisa"]
batch_size= 4
img_size = (256,256)
IMAGE_DIR= "data/Manga109_released_2023_12_07/images"
```

```python
from torch.utils.data import DataLoader
from SegDataset import MangaBubbleDataset, gather_json
from SegDataloader import manga_collate_fn, MangaBubbleDataLoader
import json
from torchvision import transforms

mask_dir= 'data/MangaSegmentation/jsons_processed'
train_json = gather_json(series_list, mask_dir)
img_size= (256,256)
TRAIN_JSON_FILE = "train.json"

with open(TRAIN_JSON_FILE, "w+") as f:
    json.dump(train_json, f, indent=2)

transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.ToTensor()])

train_loader = MangaBubbleDataLoader(TRAIN_JSON_FILE, IMAGE_DIR, img_size, batch_size=batch_size, shuffle=True,num_workers=1, transform= transform)
```

```python
from ultralytics import YOLO
YOLO_MODEL_PATH = 'best.pt'
model= YOLO(YOLO_MODEL_PATH)
```

```python
from EvalSeg import EvalSeg
from tqdm import tqdm
import torch

# Khởi tạo
all_gt_masks = []
all_gt_boxes = []
pred_masks = []
pred_boxes = []

for batch_idx, (imgs, batch_gt_masks, batch_gt_bboxes) in enumerate(tqdm(train_loader.get_loader(), desc="Processing batches")):
    
    for img_idx in range(len(batch_gt_masks)):
        all_gt_masks.append(batch_gt_masks[img_idx])
        all_gt_boxes.append(batch_gt_bboxes[img_idx])
    
    results = model.predict(imgs)
    
    for p in results:
        # Boxes cho 1 image
        img_pred_boxes = []
        for pb in p.boxes.xyxy.cpu().numpy():
            img_pred_boxes.append(pb.tolist())
        
        # Mask cho 1 image - TẠO COMBINED MASK
        if p.masks is not None:
            binary_masks = p.masks.data > 0.5  # [N, H, W]
            # Combine tất cả masks thành 1 mask
            combined_mask = binary_masks.any(dim=0)  # [H, W]
            img_pred_mask = combined_mask.cpu()
        else:
            H, W = imgs.shape[2], imgs.shape[3]
            img_pred_mask = torch.zeros((H, W), dtype=torch.bool)
        
        # Thêm vào list chính
        pred_masks.append(img_pred_mask)
        pred_boxes.append(img_pred_boxes)

print(f"Total GT masks: {len(all_gt_masks)}")
print(f"Total GT boxes: {len(all_gt_boxes)}")
print(f"Total Pred masks: {len(pred_masks)}")
print(f"Total Pred boxes: {len(pred_boxes)}")

eval_seg = EvalSeg(
    gt_masks=all_gt_masks,
    gt_bboxes=all_gt_boxes,
    pred_masks=pred_masks,
    pred_bboxes=pred_boxes,
)

print("BBOX METRICS:")
print(eval_seg.eval_bbox())
print("MASK METRICS:")
print(eval_seg.eval_mask())
```
